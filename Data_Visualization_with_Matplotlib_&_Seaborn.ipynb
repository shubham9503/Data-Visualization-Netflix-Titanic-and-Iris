```python
# Write your code here
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
a=pd.read_csv("/content/netflix_titles.csv")
a
```

```python
#Identify missing values: Check if there are any missing values in the dataset and decide how to handle them. For example, 
#if the "director" column has missing values, we can either delete those rows or fill in the missing values with "unknown".
b=a.fillna({'director':'unknown','cast':'unknown','country':'unknown'})
b
```

```python
#Data Cleaning: Clean the dataset to remove any duplicates, inconsistencies or irrelevant columns. For example, 
#we can remove the "show_id" and "description" columns as they may not be relevant for our analysis.
c=b.drop(['show_id','description'],axis=1)
c
```

```python
#Data Exploration: Use Matplotlib to create visualizations that explore the data in meaningful ways. 
#For example, we can create a bar chart to show the number of movies vs TV shows available on Netflix,
#or a line chart to show the trend in the number of new titles added each year.
c.type.value_counts().plot(kind='bar')
plt.xlabel("Title")
plt.ylabel("Year")
plt.title("Number of movies Vs TV Shows")
```

```python
x=c['title'].head(10)
y=c['release_year'].head(10)
plt.rcParams.update({'font.size':6})
plt.plot(x,y)
plt.xlabel("Title")
plt.ylabel("release_year")
plt.title("Release Year vs Title Graph")
plt.show()
```

```python
c.title.value_counts()
```

```python
#Analyze Ratings: Analyze the ratings of movies and TV shows available on 
#Netflix using histograms or box plots. For example, we can create a histogram to show the distribution of ratings on Netflix.
movie_df = b[b['type'] == 'Movie'] # Movies only
tvshow_df = b[b['type'] == 'TV Show'] # TV Shows only
tvshow_rating = tvshow_df['rating'].value_counts()
movie_rating = movie_df['rating'].value_counts()
# Plotting the data
fig = plt.figure(figsize=(10,8))
plt.rcParams.update({'font.size': 6})

plt.bar(movie_rating.index, movie_rating, label='Movie', color='#43aa8b')

plt.bar(tvshow_rating.index, tvshow_rating, label='TV Show', color='#f94144')
plt.xlabel('Rating', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.title('Quantity of content by Rating', fontsize=20)
plt.legend()
plt.show()


```

```python
# Write your code here
import seaborn as sns
Flower=sns.load_dataset('iris')
Flower
```

```python
Flower.head(10)
```

```python
Flower.describe()
```

```python
sns.scatterplot(x='sepal_length',y='sepal_width',data=Flower, label='Sepal')
sns.scatterplot(x='petal_length',y='petal_width',data=Flower, label='Petal')
plt.xlabel("Sepal and Petal Length")
plt.ylabel("Sepal and Petal Width")
plt.show()
```

```python
plt.figure(figsize=(12,12))
plt.rcParams.update({'font.size':6})
sns.boxplot(x='sepal_length',y='sepal_width',data=Flower)
sns.boxplot(x='petal_length',y='petal_width',data=Flower)
plt.xlabel("Sepal Length and Petal Length")
plt.ylabel("Sepal Width and Petal Width")
plt.show()
```

```python
#Violinplot
plt.figure(figsize=(15,10))
plt.rcParams.update({'font.size':6})
sns.violinplot(x='sepal_length',y='sepal_width',data=Flower)
sns.violinplot(x='petal_length',y='petal_width',data=Flower)
plt.xlabel("Sepal Length and Petal Length")
plt.ylabel("Sepal Width and Petal Width")
plt.show()
plt.show()
```

```python
sns.pairplot(Flower,hue='species')
```

```python
sns.heatmap(data=Flower.corr(), annot=True, cmap="winter",linewidths=0.5)
plt.show()
```
